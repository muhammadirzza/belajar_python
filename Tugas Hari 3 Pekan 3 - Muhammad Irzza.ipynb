```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(nrows=1, ncols=3, sharex=True, sharey=True, figsize=(20,8))

ax[0].plot(x, tan, color='#2f2fff', label="tanX")
ax[1].plot(x, cos, color='#ff4d4d', label="sinX")
ax[2].plot(x, sin, color='#67b367', label="cosX")

ax[0].set_title('Grafik tan x')
ax[1].set_title('Grafik sin x')
ax[2].set_title('Grafik cos x')

fig.legend(loc="upper center")

plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(sharex=True, sharey=True, figsize=(13,8))



ax.plot(x, y, color='#2f2fff')
inset_1 = ax.inset_axes([-0.2, -0.2, 0.3, 0.3])
inset_2 = ax.inset_axes([0.9, -0.2, 0.3, 0.3])
inset_3 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
inset_4 = ax.inset_axes([-0.2, 0.9, 0.3, 0.3])

inset_1.plot(x, y2, color='#ff4d4d')
inset_1.plot(x, y, color='#2f2fff')
inset_2.plot(x, y3, color='#67b367')
inset_2.plot(x, y, color='#2f2fff')
inset_3.plot(x, y4, color='#4b4b4b')
inset_3.plot(x, y, color='#2f2fff')
inset_4.plot(x, y5, color='#dada6b')
inset_4.plot(x, y, color='#2f2fff')

ax.set_title('Grafik CosX')
inset_1.set_title('Grafik CosX^2')
inset_2.set_title('Grafik CosX^3')
inset_3.set_title('Grafik CosX^4')
inset_4.set_title('Grafik CosX^5')

plt.show()

```
