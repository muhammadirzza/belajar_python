Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2

fig, ax = plt.subplots()

ax.plot(x, cos, linestyle='solid', color='#ff1e1e', label="Cosx")
ax.plot(x, cos3, linestyle='dashed', color='#68b468', label="3 Cosx")
ax.plot(x, sin, linestyle='dashdot', color='#1717ff', label="Sin(x-90)")
ax.plot(x, sin2, linestyle='None',marker='x', color='#d0d043', label="2 Sinx")

ax.set_xlabel('X Label Axis')
ax.set_ylabel('Y Label Axis')
ax.set_title('Grafik Cos Sin')

plt.legend(loc="center")


plt.show()
```
